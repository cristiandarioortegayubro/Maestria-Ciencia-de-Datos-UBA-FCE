Se estima que si las ventas fueran superiores a 900 (miles de $), el comercio podría incorporar 3 nuevos empleados. Utilizando semilla 2021, simular 5000 ventas mensuales del comercio (considerando un mes de 30 días). A partir de las mismas:

#### Habilitando librerias

````r
library(dplyr)
````

#### Seteo de la semilla

````r
set.seed(2021)
````

#### Variables

````r
VentasMensuales <- 5000
````

````r
DiasDelMes <- 30
````

**a) Determinar la distribución de las ventas mensuales (en miles de $) de dicho comercio, presentando el histograma correspondiente.**

<center>

Captura de Pantalla 2021-06-12 a la(s) 10.58.58.png

Se crea la función ventas, para poder asignarle a la misma la cantidad de dias del mes en virtud de las ventas diarias y sus probabilidades...

````r
ventas<-function(h){
  J <- sample(
    c(10,20,30,40,50),
    size=h,
    replace=T,
    prob=c(0.20,0.25,0.25,0.20,0.10))
  sum (J) }
````

Se ejecuta la funcion ventas para 30 dias...

````r
ventas(DiasDelMes)
````

Se crea un objeto "simulacion" con el valor pre establecido de 5000 ventas mensuales...

````r
simulacion <- VentasMensuales
````

````r
Y <- replicate(simulacion, ventas(DiasDelMes))
````

````r
Y[1:30]
hist(Y, col="darkorange", main = "Histograma de Ventas",ylab = "Frecuencia")
````

**b) ¿Cuál es la probabilidad empírica de que el comercio pueda incorporar 3 nuevos empleados?**

````r
VentasMes <- replicate(simulacion, ventas(DiasDelMes))
sum(VentasMes >= 900)/VentasMensuales
````

La probabilidad empirica de que el comercio pueda incorporar 3 nuevos empleados es de 0.1588
